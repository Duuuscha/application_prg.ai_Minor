# Amazon Fine Foods reviews

This dataset consists of reviews of fine foods from amazon. The data span a period of more than 10 years, including all ~500,000 reviews up to October 2012. Reviews include product and user information, ratings, and a plaintext review. We also have reviews from all other Amazon categories.

## Task description
Predicting Amazon product reviews' rating
Users can rate product on Amazon from 1 to 5 stars. The 5-star reviews contain many instances of "Great", "Good" and so on. On the other hand, the 1-star reviews use very little positive words and frequently contain warnings for potential customers. **Create a model to predict the rating of the product based on the customer review.**

```python
dict = {}
one_review = 9  #8 lines of info, one blank
N_REVIEWS = 10000

with open("finefoods.txt", "r", encoding = "latin") as file:
    for i in range(one_review*N_REVIEWS): 
        line = file.readline()
        
        if line == "\n": #Â blank lines get skipped
            continue
            
        line = line.split(": ") # dividing the key and value
        
        if line[0] in dict.keys():
            dict[line[0]]+=[line[1][:-2]] # [:-2] to stop loading end of page "\n"
        else:
            dict[line[0]]=[line[1][:-2]]
```

```python
main_df = pd.DataFrame(dict)
```

```python
text = collect_words(df,"review/text") # function defined in function.py
summary = collect_words(df,"review/summary")
```

```python
df_model = pd.DataFrame(data={"summary":summary, "text":text, "score":df["review/score"]})
```

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(df_model, test_size=0.2)

y_train = train["score"]
y_test = test["score"]
```

## Feature extraction
[sklearn](https://scikit-learn.org/stable/modules/feature_extraction.html#loading-features-from-dicts)

First I transform columns `summary` and `text` one by one then stack them together. I have tried just use one of them before but combination yields the best results. 

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer

count_vect = CountVectorizer()
X_train1 = count_vect.fit_transform(train["summary"])
X_test1 = count_vect.transform(test["summary"])

X_train2 = count_vect.fit_transform(train["text"])
X_test2 = count_vect.transform(test["text"])



tfidf_transformer = TfidfTransformer()
X_train_tfidf1 = tfidf_transformer.fit_transform(X_train1)
X_test_tfidf1 = tfidf_transformer.transform(X_test1)
X_train_tfidf2 = tfidf_transformer.fit_transform(X_train2)
X_test_tfidf2 = tfidf_transformer.transform(X_test2)


prediction={}
```

```python
from scipy.sparse import hstack
X_train_tfidf = hstack([X_train_tfidf1, X_train_tfidf2])
X_test_tfidf = hstack([X_test_tfidf1, X_test_tfidf2])
```

###Â Model Fitting

Here I am trying couple of models that should work well with this type of problem\
[sklearn](https://scikit-learn.org/stable/auto_examples/text/plot_document_classification_20newsgroups.html#sphx-glr-auto-examples-text-plot-document-classification-20newsgroups-py)

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB().fit(X_train_tfidf, y_train)
prediction['Multinomial'] = model.predict(X_test_tfidf)
print('Multinomial','%0.3f'% (model.score(X_test_tfidf, y_test)),"% mean accuracy")
```

```python
from sklearn.naive_bayes import BernoulliNB
model = BernoulliNB().fit(X_train_tfidf, y_train)
prediction['Bernoulli'] = model.predict(X_test_tfidf)
print('Bernoulli','%0.3f'% (model.score(X_test_tfidf, y_test)),"% mean accuracy")
```

```python
from sklearn import linear_model
logreg = linear_model.LogisticRegression()
logreg.fit(X_train_tfidf, y_train)
prediction['Logistic'] = logreg.predict(X_test_tfidf)
print('Logistic','%0.3f'% (logreg.score(X_test_tfidf, y_test)),"% mean accuracy")
```

```python
from sklearn.naive_bayes import ComplementNB
model = ComplementNB().fit(X_train_tfidf, y_train)
prediction['Complement'] = model.predict(X_test_tfidf)
print('Complement','%0.3f'% (model.score(X_test_tfidf, y_test)),"% mean accuracy")
```

```python
from sklearn.linear_model import SGDClassifier
model = SGDClassifier().fit(X_train_tfidf, y_train)
prediction['SGD'] = model.predict(X_test_tfidf)
print('SGD','%0.3f'% (model.score(X_test_tfidf, y_test)),"% mean accuracy")
```

Out of the models tested Logistic Regresion and SGDClassifier had similiar best accuracy in validation score

```python
from sklearn import metrics
print("Logreg",metrics.classification_report(y_test, prediction['Logistic'], target_names = ["positive", "negative"]))
print("SGD",metrics.classification_report(y_test, prediction['SGD'], target_names = ["positive", "negative"]))
```

### Confusion matrix

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, prediction['Logistic'])

cm_normalized = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
plt.figure()
plot_confusion_matrix(cm_normalized, title='Normalized confusion matrix - Logreg')  # function defined in function.py
plt.show() 

cm = confusion_matrix(y_test, prediction['SGD'])

cm_normalized = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
plt.figure()
plot_confusion_matrix(cm_normalized, title='Normalized confusion matrix - SGD')  # function defined in function.py
plt.show()
```

### ROC curve

```python
from sklearn.metrics import roc_curve, auc

def formatt(x):
    if x == 0:
        return 0
    return 1

vfunc = np.vectorize(formatt)

cmp = 0
colors = ['b', 'g', 'y', 'm', 'k']
for model, predicted in prediction.items():
    false_positive_rate, true_positive_rate, thresholds = roc_curve(y_test.map(formatt), vfunc(predicted))
    roc_auc = auc(false_positive_rate, true_positive_rate)
    plt.plot(false_positive_rate, true_positive_rate, colors[cmp], label='%s: AUC %0.2f'% (model,roc_auc))
    cmp += 1

plt.title('Classifiers comparaison with ROC')
plt.legend(loc='lower right')
plt.plot([0,1],[0,1],'r--')
plt.xlim([-0.1,1.2])
plt.ylim([-0.1,1.2])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```
